```python
df = pd.read_csv('URL dataset.csv')
df
```

```python
# Pisahkan data menjadi 1/2. stratified, urut indeks
df_legit = df[df['type'] == 'legitimate']
df_scam = df[df['type'] != 'legitimate']

# Ambil 1/2 dari masing-masing, tanpa pengacakan
df_legit_half = df_legit.iloc[:len(df_legit) // 2]
df_scam_half = df_scam.iloc[:len(df_scam) // 2]
# Gabungkan kembali
df_combined = pd.concat([df_legit_half, df_scam_half], ignore_index=False)
df_combined
```

```python
# Ambil data sisanya, simpan dalam file terpisah
df_remaining = pd.concat([df_legit.iloc[len(df_legit) // 2:], df_scam.iloc[len(df_scam) // 2:]], ignore_index=False)
df_remaining
```

```python
# Simpan kedua DataFrame ke file CSV
df_combined.to_csv('df_combined_Fahmi.csv', index=True)
df_remaining.to_csv('df_remaining_Chandra.csv', index=True)
```
